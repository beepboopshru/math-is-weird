Basic Vector Field plotting
1) Write a python program using  numpy and matplotlib to plot the vector field for the following functions F(x,y)=(-y,x). Plot the vector field in the range -5<x<5 and -5<y<5>>

```python
import numpy as np
import matplotlib.pyplot as plt

x,y=np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))

u=-y
v=x

plt.quiver(x,y,u,v)
plt.title('Vector field x,y = -y,x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Modify the above program to plot the normalized vector field. Normalize the vectors so that all arrows have the same length but preserver the direction 

This code calculates the magnitude of each each vector using the formula magnitude=np.sqrt(u^2_v^2) then normalizes u and v by dividing them by the magnitude. this ensures that all vectors have a length of 1 while preseving original directions


```python
x,y=np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))

u=-y
v=x
magnitude=np.sqrt(u**2+v**2)
u=u/magnitude
v=v/magnitude
plt.quiver(x,y,v,u)
plt.title("normalized field vectors")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x,y=np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))

u=x-y
v=x+y

plt.quiver(x,y,u,v)
plt.title('Vector field x,y = x,y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
