```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
t = np.linspace(0, 4*np.pi, 1000)

x = np.sin(3*t) * np.cos(t)
y = np.sin(3*t) * np.sin(t)
z = t
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='r(t) = (sin(3t)cos(t), sin(3t)sin(t), t)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Question 1')
ax.legend()

plt.show()
```

```python
def r(t):
    x = 2 * np.cos(t)
    y = 2 * np.sin(t)
    z = 5 * np.cos(2 * t)
    return np.array([x, y, z])

def v(t):
    x = -2 * np.sin(t)
    y = 2 * np.cos(t)
    z = -10 * np.sin(2 * t)
    return np.array([x, y, z])

def a(t):
    x = -2 * np.cos(t)
    y = -2 * np.sin(t)
    z = -20 * np.cos(2 * t)
    return np.array([x, y, z])

def speed(t):
    return np.linalg.norm(v(t))

t_val = 7 * np.pi / 4

# Position, velocity, and acceleration at t = 7/4
r_val = r(t_val)
v_val = v(t_val)
a_val = a(t_val)
speed_val = speed(t_val)

print(f"Position vector r({t_val}) = {r_val}")
print(f"Velocity vector v({t_val}) = {v_val}")
print(f"Speed at t = {t_val} = {speed_val}")
print(f"Acceleration vector a({t_val}) = {a_val}")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v_val[0], v_val[1], v_val[2], color='r', label='v(7/4)')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Velocity Vector v(7/4)')
ax.legend()

plt.show()
```

```python
def r(t):
    x = t + 1
    y = t**2 - 1
    z = 2 * t
    return np.array([x, y, z])

def v(t):
    x = 1
    y = 2 * t
    z = 2
    return np.array([x, y, z])

def a(t):
    x = 0
    y = 2
    z = 0
    return np.array([x, y, z])

def speed(t):
    return np.linalg.norm(v(t))

def direction(t):
    return v(t) / np.linalg.norm(v(t))


t_val = 1
v_val = v(t_val)
a_val = a(t_val)
speed_val = speed(t_val)
direction_val = direction(t_val)

print(f"Velocity vector v({t_val}) = {v_val}")
print(f"Acceleration vector a({t_val}) = {a_val}")
print(f"Speed at t = {t_val} = {speed_val}")
print(f"Direction of motion at t = {t_val} = {direction_val}")
```
